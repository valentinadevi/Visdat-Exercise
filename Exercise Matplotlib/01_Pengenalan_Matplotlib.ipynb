# **Pengenalan Matplotlib**

- Matplotlib adalah  salah satu module dalam pemrograman Python yang dapat digunakan untuk memvisualisasikan data dalam bentuk grafis
- Dengan Matplotlib, graps akan digambarkan pada suatu **figure**, yang di dalamnya dapat terdiri satu atau lebih axes

# Import modules

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Membuat plotting sederhana

```python
# Persiapan sample data
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

# Membuat figure dan axes
fig, ax = plt.subplots()

# Melakukan plotting data pada axes
ax.plot(x, y)
```

**Alternatif, kita bisa memanfaatkan method plot() pada pyplot untuk melakukan ploting sederhana**

```python
# persiapan sample data
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

plt.plot(x, y)
```

# Mengenal Object Oriented Style untuk melakukan plotting

Secara eksplisit membuat figures beserta axes, dan memanggil methods dari keduanya.

```python
x = np.linspace(0, 2, 100) # (p1, p2, p3) merepresentasikan sebagai (starting_point, n_point, jumlahData_point)
x
```

```python
fig, ax = plt.subplots()  # Membuat sebuah objek figure dan sebuah objek axes

# Melakukan plotting tiga variant data pada axes
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

# Menyertakan label pada sumbu x dan y serta title pada axes
ax.set_xlabel('x_label')
ax.set_ylabel('y_label')
ax.set_title('Simple Plot')

# Menyertakan legend untuk membedakan tiga variant data
ax.legend()
```

# **Mengenal pyplot Style untuk melakukan plotting**

**Mengandalkan pyplot untuk membuat dan mengelola figures dan axes, serta menggunakan fungsi pada pyplot untuk melakukan plotting**

```python
x = np.linspace(0, 2, 100) 
x
```

```python
# Melakukan plotting tiga variant data
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

# Menyertakan label pada sumbu x dan y serta title pada plot
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.title('Simple Plot')

# Menyertakan legend untuk membedakan tiga variant data
plt.legend()
```
