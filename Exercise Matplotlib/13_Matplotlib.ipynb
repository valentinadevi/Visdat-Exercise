## Belajar 3D Plot | 3D Chart

### Import Modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## 3D Axes
- Pertama buat objek figure dan objek axes.
- Namun untuk membuat axes 3D, kita perlu menyertakan parameter projection yang kita beri nilai '3d'.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

plt.show()
```

## 3D Line Plot
- Buat 3 value (X, Y, Z).
- Buat 3D plotnya.
- Berbeda dengan line plot sebelumnya, disini kita memasukkan 3 value pada ax.plot() yaitu koordinat x, y, dan z.

```python
z = np.linspace(0, 1, 100) # (Start, End, Amount)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x, y, z, color='red')

ax.set(title='Contoh 3D Line Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')

plt.show()
```

## 3D Scatter Plot
- Buat 3 value (X, Y, Z).
- Buat 3D plotnya.
- Berbeda dengan scatter plot sebelumnya, disini kita memasukkan 3 value pada ax.scatter() yaitu koordinat x, y, dan z.

```python
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(x, y, z, color='green')

ax.set(title='Contoh 3D Scatter Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')

plt.show()
```
