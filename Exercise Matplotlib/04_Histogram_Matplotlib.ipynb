# **Belajar Histogram**

# Import Modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

# Sample Data

- Nilai 10000 pada fungsi rand() akan merepresentasikan jumlah data point yang dihasilkan.

```python
# Parameter distribusi normal
mu, sigma = 100, 15 

# Menghasilkan 10.000 sampel dari distribusi normal
x = mu + sigma * np.random.randn(10000)

# Menampilkan data
x
```

```python
# Menampilkan bentuk (shape) dari array x
x.shape 
```

# Histogram dengan pyplot Style

- Untuk membuat histogram cukup gunakan fungsi plt.hist(data set, bins)
- Parameter bins berarti dari sekelompok n_data yang dimiliki akan kita bagi menjadi n_bins data.
- Untuk menampilkan teks dalam histogram bisa menggunakan plt.text(koordinat_x, koordinat_y, 'text').

```python
# Membuat Histogram
plt.hist(x, 
         bins=50,           # Jumlah bins pada histogram
         facecolor='g',     # Warna batang histogram (hijau)
         alpha=0.50)        # Tingkat transparansi batang histogram (50%)

# Menambahkan label pada sumbu x dan y
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

# Menambahkan judul pada Histogram
plt.title('Contoh Histogram')

# Menambahkan teks pada plot
plt.text(45, 500, '$\mu=100,\ \sigma=15$')  # Menampilkan teks dengan rata-rata dan deviasi standar

# Menampilkan grid pada plot
plt.grid()

# Menampilkan Histogram
plt.show()
```

# Histogram dengan OO Style

- Gunakan minimal 2 variabel untuk menampung objek figure dan axes jika menggunakan OO Style.
- Untuk menampilkan teks dalam histogram bisa menggunakan ax.text(koordinat_x, koordinat_y, 'text').

```python
# Membuat subplot dengan objek axes (ax)
fig, ax = plt.subplots()

# Membuat Histogram pada objek axes
ax.hist(x, 
        bins=50,            # Jumlah bins pada histogram
        facecolor='b',      # Warna batang histogram (biru)
        alpha=0.25)         # Tingkat transparansi batang histogram (25%)

# Menentukan label pada sumbu x dan y
ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')

# Menentukan judul pada Histogram
ax.set_title('Contoh Histogram')

# Menambahkan teks pada plot
ax.text(45, 500, '$\mu=100,\ \sigma=15$')  # Menampilkan teks dengan rata-rata dan deviasi standar

# Menampilkan grid pada plot
ax.grid()

# Menampilkan Histogram
plt.show()
```
