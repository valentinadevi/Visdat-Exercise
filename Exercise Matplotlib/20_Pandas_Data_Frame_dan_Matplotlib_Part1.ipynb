### Sample Dataset
- Membuat dataset dengan menggunakan dictionary yang terdiri dari 6 kolom.
- Dataset ini sengaja dibuat untuk memudahkan pemberian contoh pada setiap plot.
- Kolom n_anak merepresentasikan jumlah anak, sedangkan kolom n_pets merepresentasikan jumlah hewan peliharaan.

```python
df = pd.DataFrame({
    'nama': ['cecep', 'karti', 'bejo', 'tejo', 'asep', 'wati', 'amat'],
    'usia': [23, 78, 22, 19, 45, 33, 20],
    'gender': ['L', 'P', 'L', 'L', 'L', 'P', 'L'],
    'propinsi':
    ['jabar', 'jatim', 'jabar', 'jatim', 'jabar', 'jateng', 'jateng'],
    'n_anak': [2, 0, 0, 3, 2, 1, 4],
    'n_pets': [5, 1, 0, 5, 2, 2, 3]
})

df
```

### Scatter Plot
- Untuk dapat menggunakan scatter plot kita dapat memanfaatkan fungsi df.plot(). Di mana parameter kind diberi nilai 'scatter'.
- Salah satu parameter opsional dalam fungsi tersebut adalah color. Parameter ini digunakan untuk menentukan warna.
- Scatter plot digunakan untuk membandingkan dua buah variable seperti x dan y.
- Pada contoh kali ini kita menggunakan variabel bernama n_anak dan n_pets.
- df.plot(kind='scatter', x='NAMA_KOLOM ', y='NAMA_KOLOM ').

```python
df.plot(kind='scatter', x='n_anak', y='n_pets', color='red')

plt.title('Scatter Plot')
plt.xlabel('Jumlah Anak')
plt.ylabel('Jumlah Pets')
plt.show()
```

### Histogram
- Untuk membangun sebuah histogram, kita hanya perlu mengganti nilai parameter kind menjadi 'hist'.
- df[NAMA_KOLOM].plot(kind='hist', bins=[RENTANG]).

```python
df[['usia']].plot(kind='hist', bins=[0, 20, 40, 60, 80, 100])

plt.title('Histogram')
plt.xlabel('Usia')
plt.ylabel('Jumlah')
plt.show()
```

### Line Plot
- Untuk membangun sebuah line plot, kita hanya perlu mengganti nilai parameter kind menjadi 'line'.
- Pada contoh kali ini kita menggunakan variabel bernama nama dan n_anak.
- df.plot(kind='line', x='NAMA_KOLOM ', y='NAMA_KOLOM ').

```python
df.plot(kind='line', x='nama', y='n_anak', color='green')

plt.title('Line Plot')
plt.ylabel('Jumlah Anak')
plt.show()
```

### Multiple Line Plot
- Mirip dengan line plot.
- Untuk membangun sebuah multiple line plot, kita hanya perlu memodifikasi sedikit dengan menambahkan fungsi plt.gac().
- Fungsi plt.gac() memiliki singkatan 'get current axes'.
- Fungsi tersebut digunakan untuk menaruh 2 line plot kedalam 1 axes.
- Bisa dilihat perbandingan sebelum menggunakan fungsi tersebut dan sesudahnya dibawah ini.

```python
df.plot(kind='line', x='nama', y='n_anak', color='green')
df.plot(kind='line', x='nama', y='n_pets', color='red')

plt.title('Line Plot')
plt.ylabel('Jumlah')
plt.show()
```

```python
ax = plt.gca()

df.plot(kind='line', x='nama', y='n_anak', color='green', ax=ax)
df.plot(kind='line', x='nama', y='n_pets', color='red', ax=ax)

plt.title('Line Plot')
plt.ylabel('Jumlah')
plt.show()
```

### Bar Plot
- Untuk membangun sebuah bar plot, kita hanya perlu mengganti nilai parameter kind menjadi 'bar'.
- Pada contoh kali ini kita menggunakan variabel bernama nama dan usia.
- df.plot(kind='bar', x='NAMA_KOLOM ', y='NAMA_KOLOM ').
- Cara lain untuk memanfaatkan plot ini adalah dengan menggunakan fungsi group.

```python
df.plot(kind='bar', x='nama', y='usia')

plt.title('Bar Plot')
plt.ylabel('Usia')
plt.show()
```

```python
df.groupby('propinsi')['nama'].count().plot(kind='bar')

plt.title('Bar Plot')
plt.ylabel('Jumlah')
plt.show()
```

### Stacked bar plot dengan dua level group by
- Mirip dengan bar plot.
- Untuk membangun sebuah stacked bar plot, kita hanya perlu memodifikasi sedikit dengan menambahkan parameter stacked yang diberi nilai True.
- Fungsi plt.gac() memiliki singkatan 'get current axes'.
- Fungsi tersebut digunakan untuk menumpuk bar plot diatas bar plot sebelumnya.
- Bisa dilihat perbedaan peletakan kolom pada stack dibawah ini.

```python
df.groupby(['propinsi', 'gender']).size().unstack().plot(kind='bar',
                                                         stacked=True)

plt.title('Bar Plot')
plt.ylabel('Jumlah')
plt.show()
```

```python
df.groupby(['gender', 'propinsi']).size().unstack().plot(kind='bar',
                                                         stacked=True)

plt.title('Stacked Bar Plot')
plt.ylabel('Jumlah')
plt.show()
```
