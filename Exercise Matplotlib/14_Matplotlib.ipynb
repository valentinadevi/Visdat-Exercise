### Import modules

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Sample dataset
- Untuk menghasilkan data area kita dapat menggunakan fungsi np.meshgrid().
- Nilai yang ditampung X, Y, dan Z adalah array 2 dimensi.

```python
n = np.linspace(-3.0, 3.0, 100) # (Start, End, Amount)

x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
(100, 100)
(100, 100)
(100, 100)
```

### Contour Plot
- Menampilkan 3 buah data kedalam 2 dimensi.
- Perbedaan warna digunakan untuk membedakan Z/ ketinggian.
- Perbedaan tersebut bisa diukur menggunakan colour bar.

```python
plt.contour(x, y, z)
plt.colorbar()

plt.title('Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

### Filled Contour Plot
- Berbeda dengan countour plot. Filled countour plot akan memenuhi areanya dengan warna.
- Letak perbedaan fungsi ada di plt.contourf() sedangkan sebelumnya kita menggunakan plt.contour().

```python
plt.contourf(x, y, z)
plt.colorbar()

plt.title('Filled Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
