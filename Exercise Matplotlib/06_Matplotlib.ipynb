# Simple Pie Plot

- Untuk membuat pie plot kita dapat menggunakan fungsi plt.pie().

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
```

```python
# Membuat Pie Plot
plt.pie(data,              # Data yang akan diplot
        labels=kategori,   # Labels (kategori) untuk setiap sektor, opsional
        autopct='%1.1f%%', # Format tampilan persentase, opsional
        startangle=90)     # Sudut awal untuk rotasi, opsional

# Menambahkan judul pada Pie Plot
plt.title('Contoh Pie Plot')

# Menampilkan Pie Plot
plt.show()
```

# Pengaturan Legend dan Color

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
```

```python
# Membuat Pie Plot dengan warna khusus
plt.pie(data,
        labels=kategori,
        colors=warna,       # Mengganti warna setiap bagian sesuai dengan urutan warna dalam daftar
        autopct='%1.1f%%',
        startangle=90)

# Menampilkan legenda
plt.legend()

# Menambahkan judul pada Pie Plot
plt.title('Contoh Pie Plot')

# Menampilkan Pie Plot
plt.show()
```

# Exploded Pie Plot

- Exploded pie plot berarti ada bagian yang ditarik keluar.
- Semakin besar explode_car semakin jauh juga potongannya.

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08] # nilai 0.0 berarti tidak ditarik keluar.
```

```python
# Membuat Exploded Pie Plot
plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,  # Mengatur jarak sektor untuk memberikan efek "terpental" pada sektor tertentu
        autopct='%1.1f%%',
        startangle=90)

# Menambahkan judul pada Exploded Pie Plot
plt.title('Contoh Exploded Pie Plot')

# Menampilkan Exploded Pie Plot
plt.show()
```
